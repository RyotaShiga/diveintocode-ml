# 問題6 分布の確認

```python
#seabornのsns.distplot()を使い目的変数の分布を表示し、「尖度」と「歪度」も算出する。
import seaborn as sns
sns.distplot(df['SalePrice'])


```

```python
var = 'SalePrice'
# Calculate Skewness and Kurtosis(Skewness=歪度、Kurtosis=尖度)
print("Skewness :", df[var].skew())
print("Kurtosis :", df[var].kurt())
```

```python
#目的変数に対して対数変換を行う。
import numpy as np
log_target = df[var].apply(np.log)
#対数変換したものに対して分布を表示し、「尖度」と「歪度」も算出する。
import numpy as np
sns.distplot(log_target)

print("Skewness :", log_target.skew())
print("Kurtosis :", log_target.kurt())
```

logスケール変換を行う前は、尖度と歪度が大きく、ポアソン分布に従っていそうな分布であったが
行った後は正規分布に従っていそうな分布に変化した。
