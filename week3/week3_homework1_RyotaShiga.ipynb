# 問題7 特徴量間の関係の確認

```python
#縦軸と横軸に特徴量を１つずつ取り、種類毎に色分けした散布図を作る
from itertools import combinations
colors = ('red','blue','green')
label_names =X.columns 
fig = plt.figure(1,figsize=(10,20))
count = 1
for p in combinations(tuple(range(X.shape[1])),2):
        ax=plt.subplot(3,2,count)
        count += 1
        #print('count',count)
        for idx,color in enumerate(colors):
            x1 = X.iloc[y.values == idx,p[0]].values
            x2 = X.iloc[y.values == idx,p[1]].values
            plt.scatter(x1,x2,color=color,label=idx)
        plt.xlabel(label_names[p[0]])
        plt.ylabel(label_names[p[1]])
        plt.legend(loc='best')
            
            
        
```
