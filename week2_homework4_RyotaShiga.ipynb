```python
def choose_target(i=3):
    dice = np.arange(1,7)
    target = np.random.choice(dice,i)
    return target
```

# 問題2 ダーツを投げて刺さった場所を記録する

```python
def create_sample(size):
    X = np.zeros((size,2))
    target_means = []
    for l in range(1,7):
            target_means.append([l,0.0])
            
    targets = choose_target(i=size)
    #print(targets)
    for idx,target in enumerate(targets):
        x = np.random.multivariate_normal(mean=target_means[target-1],
                             cov=[[0.01,0],[0,0.01]],size=1)
        X[idx,:] = x
    return targets,X
```

```python
targets,x = create_sample(1000)
for target,sample in zip(targets,x):
    print('狙った的：{} 刺さった位置:{}'.format(target,sample))
```
